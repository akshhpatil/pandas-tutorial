## pandas merge connects columns or indexes in DataFrame based on one or more keys

```python
import pandas as pd
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4],
                  'Class':[88,99,89,67]}
                )
df1
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4],
                  'Name':['akshay','manali','aditi','aksh']})
df2
```

```python
pd.merge(df1,df2, on ="ID" )
```

```python
pd.merge(df2,df1, on ="ID" )    
```

```python
pd.merge(df2,df1, on ="ID" ,how ='left' )    
```

```python
pd.merge(df2,df1, on ="ID" ,how ='right' )    
```

```python
pd.merge(df2,df1, on ="ID" ,how ='outer' )     #outer all different values
```

```python
pd.merge(df2,df1, on ="ID" ,how ='right',indicator = True )    
```

```python
df2=pd.DataFrame({'ID':[5,6,7,8],
                  'Name':['akshay','manali','aditi','aksh']})
df2
```

```python
pd.merge(df1,df2, left_index =True ,right_index = True )  
```
