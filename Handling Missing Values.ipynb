## Handling missing values in python 

##### na_values,na_filter ,isnull,notnull

```python
import pandas as pd 
```

```python
#NaN means not a number

demo=pd.read_csv(r'C:\Users\admin\Desktop\book1.csv' ,na_values ='not avalible')
demo
```

```python
#NaN means not a number

demo=pd.read_csv(r'C:\Users\admin\Desktop\book1.csv',  keep_default_na=False )
demo
```

```python
demo=pd.read_csv(r'C:\Users\admin\Desktop\book1.csv' , na_filter=False )
demo
```

```python

demo=pd.read_csv(r'C:\Users\admin\Desktop\book1.csv')
demo
```

```python
demo.isnull()   # to check null values is present or not       
```

```python
demo.isnull().sum()        #find where null values 
```

```python
demo.isnull().sum().sum()      #find total no of null values 
```
