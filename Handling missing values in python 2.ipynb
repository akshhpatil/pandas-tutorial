# Handling missing values in python

# dropna()  drop null values rows 

```python
import pandas as pd 
```

```python
demo=pd.read_csv(r'C:\Users\admin\Desktop\book1.csv')
demo
```

```python
demo.dropna()
```

```python
demo.dropna(axis = 1)     #columns drop where null valuess
```

```python
demo.dropna(how = 'any')
```

```python
demo.dropna(how = 'any',axis = 1)
```

```python
demo.dropna(how = 'all')    # drop row and column where all null values  
```

### thresh

```python
demo.dropna(thresh = 1)  #its print not null values column and rows
```

```python
demo.dropna(thresh = 4) 
```
