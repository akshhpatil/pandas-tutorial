# pandas concat()

```python
import pandas as pd
```

```python
s1 = pd.Series([0,1,2])
```

```python
s1
```

```python
s2 = pd.Series([3,4,5,6,7])
```

```python
s2
```

```python
pd.concat([s1,s2])
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'name':['a','b','c','d'],
              
             })

df1
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8],
             'name':['r','y','h','n']
              
             })

df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df2,df1])
```

```python
pd.concat([df2,df1] ,axis = 1)
```

```python
pd.concat([df2,df1] ,axis = 0,ignore_index= True)
```

```python
df1 = pd.DataFrame({'ID':[1,2,3,4],
             'name':['a','b','c','d'],
              
             })

df1
```

```python
df2 = pd.DataFrame({'ID':[5,6],
             'name':['r','y']
              
             })

df2
```

```python
pd.concat([df1,df2] ,axis =1)
```

```python
pd.concat([df1,df2] ,axis =1 ,join = 'inner')
```

```python

df1 = pd.DataFrame({'ID':[1,2,3,4],
             'name':['a','b','c','d'],
              
             })

df1
```

```python
df2 = pd.DataFrame({'ID':[5,6,7,8],
             'name':['r','y','h','n']
              
             })

df2
```

```python
pd.concat([df1,df2],keys =['df1','df2'])
```

```python
pd.concat([df1,df2],keys =['df1','df2'] ,axis = 1)
```
