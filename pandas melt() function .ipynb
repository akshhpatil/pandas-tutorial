## pandas melt() function is used to transform or reshape data

```python
import pandas as pd 
```

```python
data =pd.read_csv('C:\\Users\\admin\\Desktop\\book1.csv')
```

```python
data
```

```python
pd.melt(data)
```

```python
pd.melt(data,id_vars =['Genders'])
```

```python
pd.melt(data,id_vars =['Age'])
```

```python
pd.melt(data,id_vars =['Age'], value_vars =['Genders'])
```

```python
pd.melt(data,id_vars =['Age'], value_vars =['Date'])
```

```python
pd.melt(data,id_vars =['Age'], value_vars =['Date'], var_name = 'category')
```

```python
pd.melt(data,id_vars =['Age'], value_vars =['Date'], var_name = 'category' ,value_name = 'data')
```
